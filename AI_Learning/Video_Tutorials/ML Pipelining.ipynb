```python
# Tutorial from https://youtu.be/HZ9MUzCRlzI?si=dzqp6AiTPIafXJTz
from sklearn.pipeline import Pipeline
#feature scaling
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
steps = [("standard_scalar" , StandardScaler()),
         ("classifier" , LogisticRegression())]


```

```python
steps
```

```python
pipe = Pipeline(steps)
```

```python
from sklearn import set_config

set_config(display = 'diagram')
pipe
```

```python
from sklearn.datasets import make_classification
import pandas as pd
X,y = make_classification(n_samples= 1000)

X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_valid = train_test_split(X,y, test_size = 0.33, random_state=42)
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
y_pred
```

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC

steps = [("scaling_scalars" , StandardScaler()),
         ("PCA", PCA(n_components = 3)),
         ("SVC", SVC())]

pipe2 = Pipeline(steps)
# pipe2['scaling_scalars'].fit_transform(X_train) to do one by one individually
pipe2.fit(X_train,y_train)
```

```python
pipe2.predict(X_test)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
#numerical processing pipeline
numeric_processor = Pipeline(
    steps = [('imputation_mean', SimpleImputer(missing_values = np.nan, strategy='mean')),
             ("scaler",StandardScaler())]
)
numeric_processor
```

```python
from sklearn.preprocessing import OneHotEncoder
#categorical processing pipeline
categorical_processor = Pipeline(
    steps = [('imputation_constant', SimpleImputer(fill_value="missing", strategy='constant')),
             ("onehot",OneHotEncoder(handle_unknown="ignore"))]
)
categorical_processor
```

```python
from sklearn.compose import  ColumnTransformer

preprocessor = ColumnTransformer(

    [('categorical', categorical_processor, ['gender' ,'City']),
     ('numeric', numeric_processor, ['age' ,'height'])]

)

preprocessor
```

```python
from sklearn.pipeline import make_pipeline
pipe3 = Pipeline(steps=[('preprocessor',preprocessor),
                            ('classifer', LogisticRegression())])
pipe3
```

```python
pipe3.fit(X_train, y_train)
```
