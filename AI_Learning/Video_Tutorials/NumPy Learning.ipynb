# **Linear Algebra**

```python
ares = np.ones((2,3))
venus = np.full((3,2) ,2)
print(ares)
print(venus)
print(np.matmul(ares,venus))
```

```python
certs = np.identity(3)
np.linalg.det(certs)
```

# **Statstics**

```python
therere = np.array([[1,2,3,4], [5,6,7,8]])
print(np.min(therere))
print(np.max(therere))
print(np.min(therere, axis=1))
print(np.sum(therere))
```

# **Reorganizing Arrays**

```python
before = np.array([[1,2,3,4], [5,6,7,8]])
print(before)
print(before.shape)
after = before.reshape((4,2))
print(after)
```

```python
#Vertical stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2])

```

```python
#Horizontal Stacking Vectors
h1= np.ones((2,4))
h2 = np.zeros((2,2))

np.hstack((h1,h2))
```
